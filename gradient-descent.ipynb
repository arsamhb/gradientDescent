```python
from mpl_toolkits import mplot3d      # tamamie in 4
%matplotlib inline                    
import numpy as np                    # rasme nemoodar 
import matplotlib.pyplot as plt       # add shodeand

from sympy import *                   # baraye moshtagh in ro add mikonim
```

```python
def f(x, y):
    return ((-1*((x)**2))-(1*((y)**2))+(6*(x))+(4*(y)+5)) #----> inja bayad avaz shavad dar soorate niaz @@@@

x = np.linspace(1, 10) # dar in do khat taeen mikonim che
y = np.linspace(1, 10) # hodoodi az in dastan dar figure namayesh dade shavad

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z');
```
